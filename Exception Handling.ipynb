# **Exception Handling in Python**

#### **What is Exception Handling ?**

- Exception handling is a mechanism used to handle runtime errors without stopping the program execution.

#### **What is Exception ?**

- An exception is an error that happens while the program is running.
- It interrupts the normal flow of the program.

#### **Difference between Error and Exception**

- Errors and exceptions are both issues in a program, but they differ in severity and handling.
- **Error**: Issues in the program logic such as SyntaxError, etc. It occurs at compile time.
- **Exception**: Problems that occur at runtime and can be managed using exception handling (e.g., invalid input, missing files).

```python
# Syntax Error (Error)
print("Hello world"  # Missing closing parenthesis

# ZeroDivisionError (Exception)
n = 10
res = n / 0
```

#### **What happens without Exception Handling?**


- Without exception handling, if an error occurs during program execution, the program will terminate immediately. The remaining lines of code will not execute.

- Exception handling helps prevent these problems by handling errors properly and allowing the program to continue execution safely.

     For example : if a program tries to divide a number by zero, Python raises a ZeroDivisionError and stops the program execution completely.


#### **Without Exception Handling**

```python
num = 10 / 0

print("Program completed")
```

##### **Note**

- Above program raises ZeroDivisionError and stops the program, the remaining line of code will not execute.

#### **With Exception Handling**

```python
try:
    num = 10 / 0

except Exception as e:
    print("Error occurred:", e)

print("Program completed")
```

- This program uses exception handling to prevent the application from crashing when division by zero occurs.

- The risky code is written inside the try block. Since dividing by zero is not allowed, Python raises a ZeroDivisionError. The except block catches this error and displays a user-friendly message instead of stopping the program.

- After handling the error, the program continues execution and prints "Program completed".

#### **`as` Keyword in Exception Handling :**

- The `as` keyword is used to store the exception message in a variable.

#### **Basic Structure of Exception Handling**

```python
try:
    # risky code
except:
    # error handling code
```

#### **Blocks Used in Exception Handling**

- try
- except
- else
- finally

**try**

- The try block is used to write the code that may generate an exception during program execution.

**except**

- The except block catches and handles the exception so that the program does not crash.

**else**

- The else block runs only if the try block executes successfully without any exception.

**finally**

- The finally block is used for cleanup operations and always executes regardless of whether an exception occurs or not.

#### **When to use `else` and `finally` Blocks in Exception Handling**

- Use the else block when you want to execute code only if no exception occurs in the try block.
- Use the finally block when some code must execute in all situations, whether an exception occurs or not.

Example :

```python
try:
    amount = int(input("Enter withdrawal amount: "))

    if amount > 5000:
        raise ValueError

except ValueError:
    print("Insufficient balance")

else:
    print("Transaction successful")

finally:
    print("Thank you for using the ATM")
```

#### **Types of Exceptions in Python**

##### **There are two major types:**

1) Built-in Exceptions
2) User-defined Exceptions

 **Built-in Exceptions**

- ZeroDivisionError
- ValueError
- TypeError
- IndexError
- KeyError
- FileNotFoundError

1) **ZeroDivisionError**
     
     - when number is divided by zero

```python
try:
    print(10 / 0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

2. **ValueError**

     - Occurs when wrong value type is given.

```python
try:
    num = int("abc")
except ValueError:
    print("Invalid number")
```

3. **TypeError**

     - A TypeError occurs when an operation is performed on incompatible data types.

```python
try:
    print("10" + 5)
except TypeError:
    print("Different data types")
```

4. **IndexError**
     
     - An IndexError occurs when trying to access an index that does not exist in a list or sequence.

```python
try:
    lst = [1, 2]
    print(lst[5])
except IndexError:
    print("Index not found")
```

5. **KeyError**

     - A KeyError occurs when trying to access a dictionary key that does not exist.

```python
try:
    d = {"name": "Prakash"}
    print(d["age"])
except KeyError:
    print("Key not found")
```

6. **FileNotFoundError**

     - A FileNotFoundError occurs when Python cannot find the specified file.

```python
try:
    open("data.txt")
except FileNotFoundError:
    print("File does not exist")
```

**User-Defined Exceptions**

- User-defined exceptions are custom exceptions created by programmers based on application requirements.
- These exceptions are created using classes.

```python
class InvalidAgeError(Exception):
    pass

age = -5

if age < 0:
    raise InvalidAgeError(
        "Age cannot be negative"
    )
```

#### **Different Ways to handle Exceptions**

- **Using Generic Exception**

    - Used to catch almost all exceptions.

```python
try:
    print(10 / 0)

except Exception as e:
    print(e)
```

- **Using Built-in Exception**
 
   - Built-in exceptions are predefined exceptions provided by Python to handle common runtime errors such as division by zero, invalid input, file not found, and invalid index access.

```python
try:
    num = int(input("Enter number: "))
    print(10 / num)

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

- **Raising Exceptions Manually**

   - Used to create exceptions manually using `raise`.


```python
age = -1

if age < 0:
    raise ValueError(
        "Age cannot be negative"
    )
```

- **User-Defined Exception Handling**

    - Used to create custom exceptions for application-specific rules.

```python
class InvalidMarksError(Exception):
    pass

try:
    marks = -10

    if marks < 0:
        raise InvalidMarksError(
            "Invalid marks"
        )

except InvalidMarksError as e:
    print(e)
```
