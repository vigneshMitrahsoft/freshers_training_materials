### What is data structure?

A Data Structure is a way of organizing, storing, and managing data in a computer so that it can be accessed and modified efficiently.

### Why Do We Need Data Structures?

#### Without proper data structures:

- Data becomes difficult to manage.
- Searching becomes slow.

#### Data structures help us:

- Store data efficiently.
- Retrieve data quickly.

### Big O Notation:

Big O tells us how efficient an algorithm is as the input size grows. it is calcuated based on below

1) Execution Time (Time complexity)
2) Memory Usage (Space Complexity)


#### 1. Execution Time (Time complexity):

If we have 10,000 records and need to find a specific one, we may have to check many records before finding it. Time Complexity describes how the effort required to perform this search increases as the number of records grows.

#### 2. Memory Usage (Space complexity):

If we have 10,000 records and need to process them, the system may require additional memory to store temporary data, copies of records, or supporting data structures. Space Complexity describes how the amount of memory required by an algorithm increases as the number of records grows.

##### Common Time Complexities:

##### 1.1) Constant Time/Space O(1):

- Constant time means an operation takes the same amount of time no matter how much data is present.
- In an array or list, if we know the index of an element, we can directly access it without searching through the collection.
- Whether the array contains 10 elements or 1 million elements, accessing an element by its index takes the same amount of time.

```python
data = [1,2,3,4,5,6]
print(data[1])
```

##### 1.2) Linear Time/Space O(n):

- Linear Time means the number of operations grows in proportion to the input size (n).
- As the amount of data increases, the algorithm may need to check more elements.
- In the worst case, it may need to check all n elements.
- Linear Search is a common example of an algorithm with O(n) time complexity.

```python
search = 6
for i in range(len(data)):
    if data[i] == search:
        print(f'Value {search} and it position {i}')
        break
```

##### 1.3) Logarithmic Time/Space O(log n):

- Logarithmic Time means the algorithm reduces the search space in every step instead of checking every value.
- It eliminates a large portion of the data based on a condition.
- Because fewer elements are checked, it requires much fewer operations than Linear Search.
- It is one of the most efficient approaches for searching in sorted data.
- Binary Search is the most common example of an algorithm with O(log n) time complexity.

##### 1.4. Quadratic Time/Space O(n^2):

- Quadratic Time means the number of operations increases as the square of the input size (n × n).
- It usually occurs when one loop runs inside another loop.
- As the input size grows, the execution time increases very quickly.
- Therefore, O(n²) algorithms are considered inefficient for large datasets.

#### 2.1.1 What is an Array?

- An Array is a linear data structure used to store multiple values in a single variable.
- The values are stored in an ordered manner, and each value can be accessed using its index (position).
- In many programming languages, an array stores elements of the same data type, so it is called a homogeneous data structure.s 

for example:
    data = ["emp1", "emp2", "emp3", ..... "empn"]

#### Why Do We Need Arrays?

Without arrays, we would need to create a separate variable for each employee like below

- e1 = emp1
- e2 = emp2
- en = empn

This becomes difficult to manage when the number of employees increases.

```python
# Array

import array

numbers = array.array('i', [10, 20, 40, 50])

# Traverse an Array
for num in numbers:
    print(num)

# Insert Element
numbers.insert(2, 30)
print(numbers)

# Append Element
numbers.append(60)
print(numbers)

# Update Element
numbers[1] = 25
print(numbers)

# Delete Element by Value
numbers.remove(30)
print(numbers)

# Delete Element by Index
numbers.pop(2)
print(numbers)
```

#### 2.1.2. What is Linked list?

A Linked List is a linear data structure made up of nodes, where each node stores data and a reference to the next node, allowing efficient insertion and deletion of elements without shifting existing data.

##### Difference between Array and Linked list:

##### Array:

- An Array stores multiple values in a sequential order.
- The values are stored next to each other in contiguous memory locations.
- Each value can be accessed using its index.
- If a new value is inserted in the middle, all elements after that position must be shifted one step to the right to make space.

Example:

        Before:
            [10, 20, 30, 40]

        Insert 25 at index 2

        After:
            [10, 20, 25, 30, 40]

##### Linked List:

- A Linked List stores data as a collection of nodes.
- Each node contains a value and a reference (pointer) to the next node.
- Nodes are not stored next to each other in memory; they can be located at different memory addresses.
- To insert a new value in the middle, we only need to update the links between nodes. Existing nodes do not need to be shifted.

Example:

        Before:
            10 → 20 → 40

        Insert 30

        After:
            10 → 20 → 30 → 40

Real World Usage:

- Undo/Redo
- music player (previous/ next)

```python
class Song:
    def __init__(self, title):
        self.title = title
        self.next = None


class Playlist:
    def __init__(self):
        self.head = None

    def add_song(self, title):
        new_song = Song(title)

        if not self.head:
            self.head = new_song
            return

        current = self.head
        # print(current.title, current.next)
        while current.next:
            current = current.next

        current.next = new_song

    def display(self):
        current = self.head

        while current:
            print(current.title, end=" -> ")
            current = current.next

        print("END")


playlist = Playlist()

playlist.add_song("Shape of You")
playlist.add_song("Believer")
playlist.add_song("Perfect")

playlist.display()
```

#### 1. Singly Linked List

A Singly Linked List consists of nodes where each node stores data and a reference to the next node.
Traversal is possible only in the forward direction.

#### Real-World Usage
- Music playlist (Next song)
- Task processing queues

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node1 = Node(10)
node2 = Node(20)

node1.next = node2

print(node1.data)
print(node1.next.data)
```

#### 2. Doubly Linked List

A Doubly Linked List consists of nodes where each node stores data, a reference to the next node, and a reference to the previous node.
Traversal is possible in both forward and backward directions.

#### Real-World Usage
- Browser back and forward navigation
- Undo and redo functionality in editors

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.prev = node1
node2.next = node3
node3.next = node1

print(node1.next.data)
print(node2.prev.data)
print(node2.next.data)
```

#### 3. Circular Linked List

A Circular Linked List is a linked list where the last node points back to the first node instead of pointing to None.
This creates a continuous loop that can be traversed repeatedly.

#### Real-World Usage
- Round Robin CPU scheduling
- Multiplayer game turn management

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

node1 = Node(10)
node2 = Node(20)
node3 = Node(30)

node1.next = node2
node2.next = node3
node3.next = node1

print(node3.next.data)
```

#### Stack:

- A Stack is a linear data structure that follows the Last In, First Out (LIFO) principle. This means the last element added to the stack is the first element to be removed.
- Think of a stack of plates: you place a new plate on top, and when you need a plate, you take the top one first.

#### when we use stack?

- A stack is useful when we always need to access the most recently added item first.

- Unlike an array, where we can access elements from any position, a stack only allows insertion and removal from the top.

Stacks are commonly used to:

- Access the most recently added data quickly.
- Keep track of recent actions or changes.
- Return to a previous state when needed.

#### Real-Wold Usage:

- Undo/Redo Operations – Used in text editors, code editors, and design tools.
- Browser History – The Back button returns to the most recently visited page.
- Image Gallery – Moving back to previously viewed images.
- Code Editors – Tracking recent changes for undo operations.

```python
class BrowserHistory:
    def __init__(self):
        self.history = []

    def visit_page(self, page):
        self.history.append(page)
        print(f"Visited: {page}")

    def go_back(self):
        if len(self.history) <= 1:
            print("No previous page available")
            return

        current_page = self.history.pop()
        print(f"Leaving: {current_page}")
        print(f"Current Page: {self.history[-1]}")

    def show_history(self):
        print("History:", self.history)


browser = BrowserHistory()

browser.visit_page("Google")
browser.visit_page("YouTube")
browser.visit_page("GitHub")

browser.show_history()

browser.go_back()
browser.go_back()

browser.show_history()
```

##### Queue:

- A Queue is a linear data structure that follows the First In, First Out (FIFO) principle. This means the first element added to the queue is the first element removed.
- Think of a queue at a ticket counter. The person who joins the line first gets served first.

#### Why Do We Use a Queue?

- A queue is used when tasks need to be processed in the same order they arrive.
- Unlike a stack, which works with the most recently added item, a queue always processes the oldest item first.

#### Queues help us:

- Process requests in order.
- Manage tasks efficiently.
- Handle multiple requests without overloading the system.
Ensure fairness by serving older requests first.
Queue Operations

#### Enqueue: Add an element to the rear of the queue.

#### Dequeue: Remove an element from the front of the queue.

#### Real-World Uses of Queue

- Ticket booking systems.
- Customer support ticket management.
- Food ordering systems.
- Call center waiting systems.
- Message queues such as RabbitMQ, Kafka, and AWS SQS.
- Request processing in web applications and APIs.


```python
from queue import Queue

ticket_queue = Queue()

ticket_queue.put("Ticket #101")
ticket_queue.put("Ticket #102")
ticket_queue.put("Ticket #103")

while not ticket_queue.empty():
    ticket = ticket_queue.get()
    print(f"Resolving {ticket}")
```
