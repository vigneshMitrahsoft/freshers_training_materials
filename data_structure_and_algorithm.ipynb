### Big O Notation:

Big O tells us how efficient an algorithm is as the input size grows. it is calcuated based on below

1) Execution Time (Time complexity)
2) Memory Usage (Space Complexity)


#### 1. Execution Time (Time complexity):

If we have 10,000 records and need to find a specific one, we may have to check many records before finding it. Time Complexity describes how the effort required to perform this search increases as the number of records grows.

#### 2. Memory Usage (Space complexity):

If we have 10,000 records and need to process them, the system may require additional memory to store temporary data, copies of records, or supporting data structures. Space Complexity describes how the amount of memory required by an algorithm increases as the number of records grows.

##### Common Time Complexities:

##### 1.1) Constant Time/Space O(1):

- Constant time means an operation takes the same amount of time no matter how much data is present.
- In an array or list, if we know the index of an element, we can directly access it without searching through the collection.
- Whether the array contains 10 elements or 1 million elements, accessing an element by its index takes the same amount of time.

##### 1.2) Linear Time/Space O(n):

- Linear Time means the number of operations grows in proportion to the input size (n).
- As the amount of data increases, the algorithm may need to check more elements.
- In the worst case, it may need to check all n elements.
- Linear Search is a common example of an algorithm with O(n) time complexity.

```python
search = 6
for i in range(len(data)):
    if data[i] == search:
        print(f'Value {search} and it position {i}')
        break
```

- It eliminates a large portion of the data based on a condition.
- Because fewer elements are checked, it requires much fewer operations than Linear Search.
- It is one of the most efficient approaches for searching in sorted data.
- Binary Search is the most common example of an algorithm with O(log n) time complexity.

#### 2.1.2. What is Linked list?

A Linked List is a linear data structure made up of nodes, where each node stores data and a reference to the next node, allowing efficient insertion and deletion of elements without shifting existing data.

##### Difference between Array and Linked list:

##### Array:

- An Array stores multiple values in a sequential order.
- The values are stored next to each other in contiguous memory locations.
- Each value can be accessed using its index.
- If a new value is inserted in the middle, all elements after that position must be shifted one step to the right to make space.

Example:

        Before:
            [10, 20, 30, 40]

        Insert 25 at index 2

        After:
            [10, 20, 25, 30, 40]

##### Linked List:

- A Linked List stores data as a collection of nodes.
- Each node contains a value and a reference (pointer) to the next node.
- Nodes are not stored next to each other in memory; they can be located at different memory addresses.
- To insert a new value in the middle, we only need to update the links between nodes. Existing nodes do not need to be shifted.

Example:

        Before:
            10 → 20 → 40

        Insert 30

        After:
            10 → 20 → 30 → 40

Real World Usage:

- Undo/Redo
- music player (previous/ next)

```python
class Song:
    def __init__(self, title):
        self.title = title
        self.next = None


class Playlist:
    def __init__(self):
        self.head = None

    def add_song(self, title):
        new_song = Song(title)

        if not self.head:
            self.head = new_song
            return

        current = self.head
        # print(current.title, current.next)
        while current.next:
            current = current.next

        current.next = new_song

    def display(self):
        current = self.head

        while current:
            print(current.title, end=" -> ")
            current = current.next

        print("END")


playlist = Playlist()

playlist.add_song("Shape of You")
playlist.add_song("Believer")
playlist.add_song("Perfect")

playlist.display()
```

#### Stack:

- A Stack is a linear data structure that follows the Last In, First Out (LIFO) principle. This means the last element added to the stack is the first element to be removed.
- Think of a stack of plates: you place a new plate on top, and when you need a plate, you take the top one first.

#### when we use stack?

- A stack is useful when we always need to access the most recently added item first.

- Unlike an array, where we can access elements from any position, a stack only allows insertion and removal from the top.

Stacks are commonly used to:

- Access the most recently added data quickly.
- Keep track of recent actions or changes.
- Return to a previous state when needed.

#### Real-Wold Usage:

- Undo/Redo Operations – Used in text editors, code editors, and design tools.
- Browser History – The Back button returns to the most recently visited page.
- Image Gallery – Moving back to previously viewed images.
- Code Editors – Tracking recent changes for undo operations.

```python
class BrowserHistory:
    def __init__(self):
        self.history = []

    def visit_page(self, page):
        self.history.append(page)
        print(f"Visited: {page}")

    def go_back(self):
        if len(self.history) <= 1:
            print("No previous page available")
            return

        current_page = self.history.pop()
        print(f"Leaving: {current_page}")
        print(f"Current Page: {self.history[-1]}")

    def show_history(self):
        print("History:", self.history)


browser = BrowserHistory()

browser.visit_page("Google")
browser.visit_page("YouTube")
browser.visit_page("GitHub")

browser.show_history()

browser.go_back()
browser.go_back()

browser.show_history()
```

#### Binary search Tree:

- A Binary Search Tree (or BST) is a data structure used to organizing and storing data in a sorted manner. 
- Each node in a Binary Search Tree has at most two children, a left child and a right child, with the left child containing values less than the parent node and the right child containing values greater than the parent node. 
- This hierarchical structure allows for efficient searching, insertion and deletion operations on the data stored in the tree.

![Image 1](https://media.geeksforgeeks.org/wp-content/cdn-uploads/20221215114732/bst-21.png)

#### Real-Wolrd Usage:

- The Contact List on Your Phone
- Auto-Complete Text Predictions

```python
class Node:
    """Represents a single node within the BST."""
    def __init__(self, key):
        self.data = key
        self.left = None   # Pointer to the left child
        self.right = None  # Pointer to the right child

class BinarySearchTree:
    """Manages the operations of the tree."""
    def __init__(self):
        self.root = None

    def insert(self, key):
        """Public method to insert a value."""
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, current, key):
        """Helper method that follows the BST rule to place new data."""
        if key < current.data:
            if current.left is None:
                current.left = Node(key)
            else:
                self._insert_recursive(current.left, key)
        elif key > current.data:
            if current.right is None:
                current.right = Node(key)
            else:
                self._insert_recursive(current.right, key)
        # Note: If key == current.data, we do nothing (BSTs typically don't allow duplicates)

    def search(self, key):
        """Public method to find if a value exists in the tree."""
        return self._search_recursive(self.root, key)

    def _search_recursive(self, current, key):
        """Helper method utilizing O(log n) efficiency."""
        # Base Cases: root is null or key is present at root
        if current is None or current.data == key:
            return current is not None

        # Key is smaller than root's key, look left
        if key < current.data:
            return self._search_recursive(current.left, key)

        # Key is greater than root's key, look right
        return self._search_recursive(current.right, key)

# --- Execution Walkthrough ---
if __name__ == "__main__":
    bst = BinarySearchTree()
    
    # Building the tree
    elements = [50, 30, 70, 20, 40, 60, 80]
    for element in elements:
        bst.insert(element)
        
    print("--- BST Verification ---")
    print("Is 40 in the tree?:", bst.search(40))  # Expected: True
    print("Is 99 in the tree?:", bst.search(99))  # Expected: False
```

#### AVL Tree:

- An AVL tree has a built-in safety rule: it constantly monitors its own shape. 
- If one branch gets even slightly heavier or taller than the other, the tree automatically rotates and shifts its nodes around to stay perfectly balanced and flat. 
- This guarantees that finding any piece of data inside it is always lightning-fast.

![image 1](https://media.geeksforgeeks.org/wp-content/uploads/20240329143458/AVL.png)

#### Real-Wolrd Usage:

- Fast Lookups in Database Indexing
- Managing High-Speed In-Memory Caches

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.height = 1

class SimpleAVL:
    def get_height(self, node):
        return node.height if node else 0

    def get_balance(self, node):
        if not node:
            return 0
        # Balance = Left Height minus Right Height
        return self.get_height(node.left) - self.get_height(node.right)

    def rotate_right(self, z):
        """Pivots the nodes to balance a left-heavy line."""
        print(f"🔄 Tree is lopsided! Rotating Right around node [{z.data}]...")
        y = z.left
        
        # Move y's right child to z's left, then make z the right child of y
        z.left = y.right
        y.right = z

        # Recalculate heights for the shifted nodes
        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y  # 'y' is the new middle/parent node!

    def insert(self, root, data):
        # 1. Standard Binary Search Tree Insertion
        if not root:
            return Node(data)
        if data < root.data:
            root.left = self.insert(root.left, data)
        else:
            root.right = self.insert(root.right, data)

        # 2. Update height of the current node
        root.height = 1 + max(self.get_height(root.left), self.get_height(root.right))

        # 3. Check Balance. If it hits 2, it's too left-heavy!
        balance = self.get_balance(root)
        if balance > 1:
            return self.rotate_right(root)

        return root

    def print_tree(self, root, level=0):
        """Helper to print the tree sideways in the console."""
        if root:
            self.print_tree(root.right, level + 1)
            print("    " * level + f"[{root.data}]")
            self.print_tree(root.left, level + 1)

# --- Run the Simulation ---
avl = SimpleAVL()
root = None 

# We insert 30, then 20, then 10. (A straight line down the left)
for num in [30, 20, 10, 80, 90, 250, 100]:
    print(f"\nAdding: {num}")
    root = avl.insert(root, num)
    print("Current Tree Shape:")
    avl.print_tree(root)
```

#### Linear Search:

Linear Search is the simplest searching algorithm used to find an element in a collection of data. It checks each element one by one from the beginning until the required element is found. If the element is not present, the search continues until the end of the collection.

#### Why We Use linear search?

- Linear Search is easy to understand and implement for beginners. 
- It works on both sorted and unsorted data, making it very flexible. 
- This algorithm is suitable for small datasets where search performance is not a major concern.

#### Real-world Uasge:

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

numbers = [10, 25, 30, 45, 50]

result = linear_search(numbers, 85)

if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```

#### Binary Search

Binary Search is an efficient searching algorithm that works only on sorted data. It repeatedly divides the search space into two halves and compares the middle element with the target value. This process continues until the element is found or the search space becomes empty.

#### Why We Use binary search?

- Binary Search significantly reduces the number of comparisons required to find an element. 
- It provides much faster results than Linear Search for large datasets. 
- This algorithm is preferred when data is already sorted and quick searching is required.

#### Real-world Uasge:

- Database Indexing (B-Trees and B+ Trees)
- E-Commerce Price Filtering


```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        elif arr[mid] < target:
            left = mid + 1

        else:
            right = mid - 1

    return -1

numbers = [10, 20, 30, 40, 50, 60, 70]

result = binary_search(numbers, 85)

if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```

#### Heap:

- A Heap is a special tree-based data structure that satisfies the heap property. It is a complete binary tree, which means all levels are completely filled except possibly the last, and the last level is filled from left to right.
- In a **Max Heap**, every parent node is greater than or equal to its children, so the largest element is always at the root.
- In a **Min Heap**, every parent node is smaller than or equal to its children, so the smallest element is always at the root.
- Although a heap is logically a binary tree, it is usually stored in an array. For a node at index `i`, its left child is at `2*i + 1`, its right child is at `2*i + 2`, and its parent is at `(i - 1) // 2`.

#### Why We Use heap?

- A heap lets us access the minimum or maximum element instantly in O(1) time.
- Insertion and deletion of the priority element take only O(log n) time, which is very efficient.
- It is the most common way to implement a Priority Queue, where elements are served based on priority instead of arrival order.

#### Real-World Usage

- Priority queues in operating systems for CPU and task scheduling.
- Finding the shortest path in graphs using Dijkstra's algorithm.
- Building Heap Sort, an efficient comparison-based sorting algorithm.
- Getting the top K largest or smallest elements from large datasets.

#### Heap Operations

- **Insert:** Add a new element and move it up until the heap property is restored.
- **Peek:** Look at the root element (minimum or maximum) without removing it.
- **Extract:** Remove the root element and rearrange the heap to maintain the heap property.

```python
class MinHeap:
    def __init__(self):
        self.heap = []

    def parent(self, i):
        return (i - 1) // 2

    def left_child(self, i):
        return 2 * i + 1

    def right_child(self, i):
        return 2 * i + 2

    # Insert a new value and move it up to its correct position
    def insert(self, value):
        self.heap.append(value)
        i = len(self.heap) - 1
        while i > 0 and self.heap[self.parent(i)] > self.heap[i]:
            self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
            i = self.parent(i)

    # Look at the smallest element without removing it
    def peek(self):
        if not self.heap:
            return None
        return self.heap[0]

    # Remove and return the smallest element, then fix the heap
    def extract_min(self):
        if not self.heap:
            return None
        root = self.heap[0]
        last = self.heap.pop()
        if self.heap:
            self.heap[0] = last
            self.heapify_down(0)
        return root

    def heapify_down(self, i):
        size = len(self.heap)
        while True:
            smallest = i
            left = self.left_child(i)
            right = self.right_child(i)
            if left < size and self.heap[left] < self.heap[smallest]:
                smallest = left
            if right < size and self.heap[right] < self.heap[smallest]:
                smallest = right
            if smallest == i:
                break
            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            i = smallest


heap = MinHeap()

for number in [40, 10, 30, 5, 20]:
    heap.insert(number)

print("Heap array:", heap.heap)
print("Minimum element (peek):", heap.peek())

print("Extracted min:", heap.extract_min())
print("Heap after extraction:", heap.heap)
```

#### 1. Bubble Sort:

Bubble Sort repeatedly compares adjacent elements and swaps them if they are in the wrong order. Larger elements gradually move toward the end of the list after each pass. It is simple to understand but not efficient for large datasets.

#### why we use bubble sort?

- It is the absolute simplest sorting method to code and understand for beginners.
- It sorts the items directly inside the original list, meaning it takes up virtually zero extra computer memory.

#### Real-world uasge

- Computer Graphics & UI Visualizations
- Nearly Sorted Data Verification

![step 1](https://media.geeksforgeeks.org/wp-content/uploads/20240925153535/bubble-sort-1.webp)
![step 2](https://media.geeksforgeeks.org/wp-content/uploads/20240925153536/bubble-sort-2.webp)
![step 3](https://media.geeksforgeeks.org/wp-content/uploads/20240925153536/bubble-sort-3.webp)


```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

numbers = [64, 34, 25, 12, 22, 11, 90]

sorted_numbers = bubble_sort(numbers)

print("Sorted Array:", sorted_numbers)
```

#### Selection sort:

Selection Sort is a simple sorting algorithm that repeatedly finds the smallest element from the unsorted part of the array and places it in its correct position. After each pass, one element is permanently placed in the sorted portion of the array. This process continues until the entire array becomes sorted.

#### why we use selection sort?

- It performs fewer swaps compared to Bubble Sort, which can be beneficial when swap operations are expensive. 
- Although it is not efficient for large datasets, it works well for small collections of data.

#### Real-world uage:

```python
def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

numbers = [64, 25, 12, 22, 11]

sorted_numbers = selection_sort(numbers)

print("Sorted Array:", sorted_numbers)
```

#### Insertion Sort

Insertion Sort is a simple sorting algorithm that builds a sorted array one element at a time. It takes an element from the unsorted portion and inserts it into its correct position in the sorted portion. This process continues until all elements are arranged in the required order.

#### Why We Use insertion Sort?

- Insertion Sort is easy to implement and performs efficiently for small datasets. 
- It works particularly well when the data is already sorted or nearly sorted. 
- The algorithm requires very little extra memory because sorting is performed within the same array.

#### Real-world usage

- Online Card Games
- Continuous Sensor Streams

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr

numbers = [12, 11, 13, 5, 6]

sorted_numbers = insertion_sort(numbers)

print("Sorted Array:", sorted_numbers)
```

#### Quick Sort

Quick Sort is a highly efficient sorting algorithm that follows the Divide and Conquer approach. It selects a pivot element and partitions the array into two parts, where elements smaller than the pivot are placed on the left and larger elements are placed on the right. The same process is then applied recursively to the left and right subarrays until the entire array becomes sorted.

#### Why We Use quick sort?

- Quick Sort is one of the fastest sorting algorithms for large datasets in practical applications. 
- It reduces the number of comparisons required by repeatedly dividing the array into smaller parts. 
- Due to its high performance and efficiency, Quick Sort is widely used in software systems and programming libraries.

#### Real-world Usage

- Graphics Engines
- Commercial Analytics Software

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]

    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    print(left, middle, right)

    return quick_sort(left) + middle + quick_sort(right)


numbers = [10, 7, 8, 9, 1, 5]

sorted_numbers = quick_sort(numbers)

print("Sorted Array:", sorted_numbers)
```

#### Merge Sort

Merge Sort is an efficient sorting algorithm that follows the Divide and Conquer approach. It repeatedly divides the array into smaller halves until each subarray contains only one element. The smaller subarrays are then merged back together in sorted order to produce the final sorted array.

#### Why We Use merge sort?

- Merge Sort provides consistent performance for small and large datasets. 
- Its time complexity remains O(n log n) in all cases, making it more predictable than many other sorting algorithms. 
- It is also a stable sorting algorithm, meaning equal elements maintain their original order after sorting.

#### Real-world usage

- E-Commerce Product Filters
- Language Runtimes (Python sort() )

![step 1](https://media.geeksforgeeks.org/wp-content/uploads/20250923102849709166/arr_.webp)

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):

        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


numbers = [38, 27, 43, 3, 9, 82, 10]

sorted_numbers = merge_sort(numbers)

print("Sorted Array:", sorted_numbers)
```
