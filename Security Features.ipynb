##### **1) Authentication**

- Authentication is the process of verifying a user's identity before allowing access to an application.

*Example* :

   - When you log in to Instagram using your username and password, the application verifies your identity before giving access to your account.

##### **2) Authorization**

- Authorization is the process of deciding what an authenticated user is allowed to access or perform in an application.

*Example*
    
   - In a company application, an employee can view their profile, but only the admin can access employee salary details.

##### **3)Password Hashing**

- Password hashing is the process of converting a password into an unreadable fixed value to store it securely.

*Example*
   - When you create a password in Instagram, the application stores the hashed version of your password instead of the original password.

```python
# With Password Hashing

import hashlib

password = "admin123"

hashed = hashlib.sha256(password.encode()).hexdigest()

print(hashed)  # Password becomes a unreadable format, So even hackers get access to the database,
               # they won't be able to see the actual password to log in.
```

```python
# Without Password Hashing

password = "admin123"  # If we store the password in plain text, anyone who gets access to the database can see the actual password and log in.
```

#### **4) Encryption**

- Encryption is the process of converting readable data into an unreadable format to protect it from unauthorized access.
- Only authorized users with the correct key can read the original data.

#### **5) Decryption**

- Decryption is the process of converting encrypted unreadable data back into its original readable form using a key.

*Example*

**When you send a message in WhatsApp:**

   - The message is encrypted before sending
   - Others cannot read it during transmission
   - The receiver decrypts and reads the original message

#### **Types of Encryption**

1) Symmetric Encryption
2) Asymmetric Encryption

**Most Common Python Libraries :**
   - cryptography
   - PyCryptodome

##### **1) Symmetric Encryption**

- Same key is used for encryption and decryption

*Example :*

- In banking applications, transaction data is encrypted and decrypted using the same secret key to keep communication secure.

###### STEP 1: Server Shares Public Key


        Server  -------------------->  Banking App
                    PUBLIC_KEY


###### STEP 2: App Creates Secret Key


        Secret Key = BANK123


###### STEP 3: App Encrypts Secret Key


        BANK123  -->  X@#9Lm
        (using PUBLIC_KEY)


###### STEP 4: Encrypted Key Sent Through Internet


        Banking App  -------------------->  Server
                            X@#9Lm


###### STEP 5: Server Decrypts Secret Key


        X@#9Lm  -->  BANK123
        (using PRIVATE_KEY)


###### STEP 6: Symmetric Encryption Starts


        Banking App  <------------------>  Server

        Both use same Secret Key:
        BANK123

```python
from cryptography.fernet import Fernet

# STEP 1: Generate secret key
key = Fernet.generate_key()

# STEP 2: Create cipher object
cipher = Fernet(key)

# Original message
message = b"Hello Python"

# STEP 3: Encrypt message
encrypted_message = cipher.encrypt(message)

print("Encrypted Message:")
print(encrypted_message)

# STEP 4: Decrypt message
decrypted_message = cipher.decrypt(encrypted_message)

print("\nDecrypted Message:")
print(decrypted_message.decode())
```

##### **2) Asymmetric Encryption**

- Asymmetric encryption is a security method that uses two different keys:

        Public Key → Used to encrypt data
        Private Key → Used to decrypt data

*Example :*

**Email Encryption**

In secure email systems:
  - Sender encrypts the email using receiver’s public key
  - Receiver decrypts it using their private key

*How private key decrypts the message ?*

   - The private key decrypts the message because it is mathematically linked to the public key that encrypted the message.

```python
from cryptography.hazmat.primitives.asymmetric import rsa, padding
from cryptography.hazmat.primitives import hashes

# Generate private and public keys
private_key = rsa.generate_private_key(
    public_exponent=65537,  #65537 is a standard secure mathematical value used internally by RSA to create and use public keys safely.
    key_size=2048
)

public_key = private_key.public_key()

# Original message
message = b"Hello"

# Encrypt using public key
encrypted = public_key.encrypt(
    message,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

print("Encrypted:", encrypted)

# Decrypt using private key
decrypted = private_key.decrypt(
    encrypted,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)

print("Decrypted:", decrypted.decode())
```

#### **6) SQL Injection**

 - SQL Injection is a cyber attack where hackers insert malicious SQL code into application input fields to access or manipulate database data.

***query = f"SELECT * FROM users WHERE username='{username}' AND password='{password}'"***

Username: admin
Password: ' OR '1'='1

***SELECT * FROM users***
***WHERE username='admin'***
***AND password='' OR '1'='1';***

'1'='1' is always TRUE.
So database may allow login without correct password.
This is SQL Injection.

##### **How to prevent SQL Injection**

1) Parameterized Queries
2) ORM frameworks

***Parameterized Queries***

```python
import sqlite3

conn = sqlite3.connect("users.db")
cursor = conn.cursor()

username = "admin"
password = "1234"

cursor.execute(
    "SELECT * FROM users WHERE username=? AND password=?",
    (username, password)
)

result = cursor.fetchall()

print(result)

#The input values are treated only as data, not as SQL commands.
#Even if hacker enters:
#' OR '1'='1
#Python treats it as normal text.So, No SQL injection occurs.
```

#### **7) CSRF**

- It is an attack where a hacker tricks a logged-in user into performing unwanted actions on a website.

*Example*

   Suppose you are logged into your banking application.

   Without logging out,
   you visit a malicious website.

   That website secretly sends a request like:

   **Transfer ₹10,000 to Hacker**

   Because your browser is already logged into the bank:

   Bank thinks the request came from you
   Money transfer may happen

   This is CSRF attack.
  

##### **How to prevent CSRF ?**

- Using CSRF Token in Form/Page
- CSRF tokens prevent it by verifying genuine requests to process it.

#### **9) JWT token-Based Authentication**

- JWT authentication uses secure tokens to verify users after login without repeatedly sending usernames and passwords.

##### **Without JWT**
- Server stores many sessions
- Repeated password verification
- Poor scalability
- Slower performance

##### **With JWT**
- User logs in once
- Token used for future requests
- No need to send password repeatedly
- Stateless authentication

##### **Flow of JWT**

1) User Login
    
2) Server Generates JWT Token
    
3) Client Stores Token
    
4) Client Sends Token in Every Request
    
5) Server Verifies Token
    
6) Access Granted

#### **10) XSS (Cross-Site Scripting)**

- It is a web security attack where hackers inject malicious JavaScript code into a website.
- When other users open the page, the malicious script executes in their browser.

##### *Example*

- Suppose a website has a comment section.
- Normal user comment: 
    - Nice article!
- Hacker comment:
    - Alert script in comment section.
    - Browser runs the JavaScript code, and popup appears to website.

#### **With XSS Protection**

- It is a security protection against attacks where hackers inject malicious JavaScript code into websites.

*How it prevent attacks*

   - Input Validation
   - Output Escaping
   - Content Security Policy (CSP)  - it restricts which scripts and resources browsers are allowed to load and execute.
