## 1. Introduction

This section explains what SQLAlchemy is, why it is useful, and how it fits into Python applications that work with databases.

### What is SQLAlchemy?

- **Definition:** SQLAlchemy is a Python library that helps applications talk to databases.
- **Why it is used:** It makes working with database tables easier by using Python objects instead of writing SQL every time.
- **Supported databases:** SQLite, PostgreSQL, MySQL, MariaDB, SQL Server, Oracle, and many others.
- **SQLAlchemy Core vs ORM:**
  - SQLAlchemy Core gives you low-level database access using SQL expressions.
  - SQLAlchemy ORM lets you work with database rows as Python objects using mapping rules.

### Real-world Example

Imagine an Employee Management System where we store employee information like name, email, department, and salary. SQLAlchemy helps us keep this information in a database while using simple Python classes inside the application.

You can think of SQLAlchemy as a bridge between the application code and the database tables.

## 2. Database Connectivity

In this section, we learn how SQLAlchemy connects to a database and what an engine does.

### Why Database Connection is Needed

An application needs a connection to a database to send queries, get results, and store information. Without a connection, the code cannot access the database storage.

### What is an Engine?

- The engine is a SQLAlchemy object that manages the database connection.
- It knows how to talk to a specific database using a URL.
- The engine is like a connection manager or driver between your code and the database.

### Database URLs

A database URL tells SQLAlchemy where the database is and what kind it is. Examples:

- SQLite: `sqlite:///example.db`
- PostgreSQL: `postgresql+psycopg2://user:password@localhost/mydb`
- MySQL: `mysql+pymysql://user:password@localhost/mydb`

````python
# Install SQLAlchemy in a notebook environment if needed
# Uncomment and run the following line if SQLAlchemy is not installed:
# !pip install SQLAlchemy

from sqlalchemy import create_engine

# Create a SQLite engine that stores data in a local file
engine = create_engine('sqlite:///sqlalchemy_orm_demo.db', echo=False)
print('Engine created:', engine)
````

### Creating an Engine

The code above creates an engine for SQLite. The same idea applies to other databases by changing the URL.

Expected output explanation:
- `Engine created:` means the connection manager is ready.
- SQLAlchemy does not immediately open the database file until it runs queries.

### Real-world Example

An engine is like a receptionist at a company. The application asks the engine for a connection to the database, and the engine brings the correct connection to perform queries.

## 3. ORM (Object Relational Mapping)

This section introduces ORM and why it is useful when working with databases in Python.

### What is ORM?

- ORM stands for Object Relational Mapping.
- It means converting data between Python objects and database tables.
- In SQLAlchemy ORM, each table is usually represented by a Python class.

### Why ORM Exists

ORM exists because writing raw SQL for every database operation can be repetitive and error-prone.
It allows developers to write Python code instead of managing SQL strings directly.

### How ORM Works

ORM maps database concepts to Python concepts. Here is a simple table-to-class mapping:

- Database Table -> Python Class
- Database Row -> Python Object
- Column -> Attribute

| Database Concept | ORM Concept |
|---|---|
| Table | Class |
| Row | Object |
| Column | Attribute |

### Advantages of ORM

- Easier to work with Python objects.
- Less raw SQL in the application code.
- Helps keep database logic organized.
- Allows reuse of Python classes across the application.

### Disadvantages of ORM

- Can hide database details and make queries less explicit.
- May be slower than raw SQL for complex queries.
- Requires understanding of both Python classes and relational databases.

### Real-world Example

Suppose we have an `employees` table in a database. ORM allows us to use an `Employee` class in Python, where each instance holds data for one employee. This is easier to read than passing SQL strings around the code.

## 4. Models

Models are the core of ORM. They define how tables and columns look in Python.

### What is a Model?

- A model is a Python class that describes a database table.
- It contains information about columns and relationships.

### Why Models are Needed?

Models provide a clear structure for database tables and make it easier to insert, query, and update data using Python code.

### Creating a Model

A model usually requires a base class and column definitions. In SQLAlchemy ORM, we use `DeclarativeBase`, `__tablename__`, `mapped_column`, and `Mapped`.

````python
from sqlalchemy import Integer, String
from sqlalchemy.orm import DeclarativeBase, Mapped, mapped_column

class Base(DeclarativeBase):
    pass

class Employee(Base):
    __tablename__ = 'employees'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(100), nullable=False)
    email: Mapped[str] = mapped_column(String(150), unique=True, nullable=False)

print('Model created:', Employee.__name__)
````

### Code Example Explanation

- `DeclarativeBase` is the base class for all ORM models.
- `__tablename__` defines the database table name.
- `mapped_column` defines a column with a type and optional constraints.
- `Mapped` tells SQLAlchemy the attribute is part of the mapped model.

Expected output explanation:
- `Model created: Employee` means the ORM class is defined successfully.

## 5. Model Components

Models are built from data types, constraints, and default values.

### Data Types

Common SQLAlchemy column types include:

- `Integer`: whole numbers.
- `String`: short text.
- `Boolean`: True/False values.
- `Float`: decimal numbers.
- `Date`: calendar date.
- `DateTime`: date and time.
- `Text`: long text.

These types tell the database how to store the values.

````python
from sqlalchemy import Boolean, Date, DateTime, Float, Text
from datetime import date, datetime

class Example(Base):
    __tablename__ = 'examples'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    active: Mapped[bool] = mapped_column(Boolean, default=True)
    score: Mapped[float] = mapped_column(Float, default=0.0)
    description: Mapped[str] = mapped_column(Text, nullable=True)
    start_date: Mapped[date] = mapped_column(Date, default=date.today)
    created_at: Mapped[datetime] = mapped_column(DateTime, default=datetime.utcnow)

print('Example model defined with many data types.')
````

### Constraints

Constraints define rules for data in a table. Common constraints are:

- `primary_key=True`: uniquely identifies each row.
- `ForeignKey`: links one table to another.
- `unique=True`: ensures values are not repeated.
- `nullable=False`: requires a value in the column.
- `CheckConstraint`: enforces custom rules.

````python
from sqlalchemy import CheckConstraint, ForeignKey

class Department(Base):
    __tablename__ = 'departments'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(100), unique=True, nullable=False)

class TeamMember(Base):
    __tablename__ = 'team_members'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    department_id: Mapped[int] = mapped_column(ForeignKey('departments.id'), nullable=False)
    salary: Mapped[float] = mapped_column(Float, CheckConstraint('salary >= 0'), default=0.0)

print('Department and TeamMember models defined with constraints.')
````

### Default Values

Default values are used when a field is not provided during insert.
Example: `default=True`, `default=0`, or `default=date.today`.

## 6. Relationships

Relationships help connect models that depend on each other.

### Why Relationships are Needed?

Relationships let us access related rows easily without writing manual joins every time. They model real-world links between data.

### One-to-One Relationship

A one-to-one relationship means one row in table A links to exactly one row in table B.

**Example:** `User` and `Profile` where each user has one profile.

Text diagram:
```
User 1 <----> 1 Profile
```

**Model code:**

````python
from sqlalchemy.orm import relationship

class User(Base):
    __tablename__ = 'users'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    username: Mapped[str] = mapped_column(String(50), unique=True, nullable=False)
    profile: Mapped['Profile'] = relationship('Profile', back_populates='user', uselist=False)

class Profile(Base):
    __tablename__ = 'profiles'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    user_id: Mapped[int] = mapped_column(ForeignKey('users.id'), nullable=False)
    bio: Mapped[str] = mapped_column(String(200), nullable=True)
    user: Mapped[User] = relationship('User', back_populates='profile')

print('One-to-one relationship defined between User and Profile.')
````

### One-to-Many Relationship

One department can have many employees.

Text diagram:
```
Department 1 <----> * Employee
```

**Model code:**

````python
class Department(Base):
    __tablename__ = 'departments'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(100), nullable=False)
    employees: Mapped[list['Employee']] = relationship('Employee', back_populates='department')

class Employee(Base):
    __tablename__ = 'employees'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(100), nullable=False)
    department_id: Mapped[int] = mapped_column(ForeignKey('departments.id'))
    department: Mapped[Department] = relationship('Department', back_populates='employees')

print('One-to-many relationship defined between Department and Employee.')
````

### Many-to-Many Relationship

Many students can take many courses.

Text diagram:
```
Student * <----> * Course
```

**Model code:**

````python
from sqlalchemy import Table

association_table = Table(
    'student_courses',
    Base.metadata,
    mapped_column('student_id', ForeignKey('students.id'), primary_key=True),
    mapped_column('course_id', ForeignKey('courses.id'), primary_key=True)
)

class Student(Base):
    __tablename__ = 'students'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    name: Mapped[str] = mapped_column(String(100), nullable=False)
    courses: Mapped[list['Course']] = relationship('Course', secondary=association_table, back_populates='students')

class Course(Base):
    __tablename__ = 'courses'

    id: Mapped[int] = mapped_column(Integer, primary_key=True)
    title: Mapped[str] = mapped_column(String(100), nullable=False)
    students: Mapped[list[Student]] = relationship('Student', secondary=association_table, back_populates='courses')

print('Many-to-many relationship defined between Student and Course.')
````

## 7. Sessions

Sessions are used to organize changes before they are saved to the database.

### What is a Session?

A session is a temporary workspace where SQLAlchemy tracks objects and changes before sending them to the database.

### Why Session is Needed?

The session keeps a list of changes, helps group operations, and allows commit or rollback of work as a unit.
A good analogy is a shopping cart or a draft document.

### Session Methods

- `add()`: add a new object to the session.
- `commit()`: save all changes to the database.
- `rollback()`: undo changes if something goes wrong.
- `refresh()`: reload an object from the database.
- `close()`: close the session and release resources.

````python
from sqlalchemy.orm import Session

with Session(engine) as session:
    employee = Employee(name='Sara Lee', email='sara@example.com', department_id=1)
    session.add(employee)
    session.commit()
    print('Employee added and committed.')

    # Refresh the object from the database in case defaults were added
    session.refresh(employee)
    print('Employee id after commit:', employee.id)
````

## 8. CRUD Operations

CRUD stands for Create, Read, Update, Delete. These are the basic operations for working with data.

### Create

SQLAlchemy code:

````python
with Session(engine) as session:
    new_employee = Employee(name='Amit Patel', email='amit.patel@example.com', department_id=1)
    session.add(new_employee)
    session.commit()
    print('Created employee:', new_employee.name)
````

Equivalent SQL query:
```
INSERT INTO employees (name, email, department_id) VALUES ('Amit Patel', 'amit.patel@example.com', 1);
```

### Update

SQLAlchemy code:

````python
with Session(engine) as session:
    employee = session.scalars(select(Employee).where(Employee.name == 'Amit Patel')).first()
    if employee:
        employee.email = 'amit.updated@example.com'
        session.commit()
        print('Updated email for', employee.name)
````

Equivalent SQL query:
```
UPDATE employees SET email = 'amit.updated@example.com' WHERE name = 'Amit Patel';
```

## 9. Querying Data

Learn how to select and filter data using SQLAlchemy queries.

### select()

Use `select()` to build a query. It is equivalent to `SELECT` in SQL.

### where()

Use `.where()` to filter rows. This is equivalent to `WHERE` in SQL.

### order_by()

Use `.order_by()` to sort results. This is equivalent to `ORDER BY` in SQL.

### limit()

Use `.limit()` to limit the number of rows returned. This is equivalent to `LIMIT` in SQL.

### joins()

Use `join()` to combine related tables. This is equivalent to `JOIN` in SQL.

````python
with Session(engine) as session:
    statement = select(Employee).where(Employee.id > 0).order_by(Employee.name).limit(5)
    employees = session.scalars(statement).all()
    print('Top employees:')
    for emp in employees:
        print(emp.id, emp.name, emp.email)
````

Equivalent SQL query:
```
SELECT * FROM employees WHERE id > 0 ORDER BY name LIMIT 5;
```

## 12. Limitations of SQLAlchemy

Sometimes raw SQL is better than ORM. Here are situations where ORM may not be ideal.

### Complex Database-Specific Features

When your application uses database-specific functions or syntax, raw SQL may be easier to control.

Example: Stored procedures written in the database often work better with direct SQL calls.

````python
sql = 'SELECT COUNT(*) FROM employees WHERE salary > :min_salary'
with engine.connect() as conn:
    result = conn.execute(text(sql), {'min_salary': 50000})
    print('Raw SQL count result:', result.scalar())
````

### Stored Procedures

If your database uses stored procedures for important logic, SQLAlchemy ORM may not be the best choice for calling them. Raw SQL or database-specific APIs can be more direct.

### Vendor-Specific Optimizations

Some databases have optimization features that are hard to express with ORM. Raw SQL lets you tune queries exactly for the database engine.

### Large Reporting Queries

For reports that process large amounts of data, raw SQL is often faster and easier to maintain than ORM-generated queries.

Tradeoff: ORM is simpler for normal CRUD operations, but raw SQL may be needed for heavy data work.

### Bulk Data Processing

Bulk inserts or updates may be better handled with raw SQL or SQLAlchemy Core for performance reasons.

Tradeoff: ORM works well for small to medium workloads, but not always for very large batches.
