### Embeddings
The Problem

Computers understand numbers.

Humans understand words.

```
Human:
"dog"

Computer:
???
```
An LLM cannot directly work with the meaning of the word "dog".

It needs a numerical representation.

What is an Embedding?

An embedding is a list of numbers that represents the meaning of a piece of text.

```
dog

↓

[0.12, -0.43, 0.91, 0.22, ...]
```
This list is called a vector.

#### Why do we need Embeddings?

Suppose we have:
```
dog
puppy
car
```
Humans know:
```
dog ≈ puppy

dog ≠ car
```
We want computers to know that too.

Embeddings make semantically similar words have similar vectors.

Example:
```
dog
↓

[0.9, 0.8, 0.7]

puppy
↓

[0.88, 0.81, 0.69]

car
↓

[0.1, 0.2, 0.9]
```
Notice:
```
dog and puppy

are close together

dog and car

are far apart
```
Except:

A hash only identifies.

An embedding captures meaning.
```
dog
↓

Embedding

↓

Meaning preserved
```

### Vector
A vector is just:
```
[1, 2, 3]
```
or
```
[100]
```
That's it.

A vector is simply an ordered list of numbers.

Imagine you create your own mapping:
```
words = {
    "dog": 1,
    "cat": 2,
    "python": 3
}
```
Then:
```
"dog"
```
becomes
```
1
```
and
```
"python"
```
becomes
```
3
```
This is already text → number.

Embeddings are a smarter version

Instead of:
```
"dog" -> 1
"cat" -> 2
"car" -> 3
```
an embedding model does:
```
"dog" -> [0.8, 0.2, 0.9]
"puppy" -> [0.7, 0.3, 0.8]
"car" -> [0.1, 0.9, 0.2]
```
Notice something?
```
dog
puppy
```
have similar numbers.

Because they have similar meanings.

### Similarity Search
Imagine you're building a Django application.

You store documents:
```
Document 1:
Django is a Python web framework.

Document 2:
FastAPI is used for building APIs.

Document 3:
Dogs are loyal pets.
```
Before storing them, you generate embeddings.
```
Document 1
↓
Embedding Model
↓
[0.81, 0.22, 0.44, ...]
```
```
Document 2
↓
Embedding Model
↓
[0.79, 0.18, 0.41, ...]
```
```
Document 3
↓
Embedding Model
↓
[0.12, 0.91, 0.73, ...]
```
User asks a question
```
How do I create REST APIs in Python?
```
We also convert this question into a vector.
```
Question
↓
Embedding Model
↓
[0.80, 0.20, 0.43, ...]
```
Now we have:
```
Question Vector
```
and
```
Document Vectors
```
#### Similarity Search
Now the system asks:

```
document vector is most similar to the question vector?
```
Something like:
```
Question Vector
      ↓

Document 1 → similarity = 0.85

Document 2 → similarity = 0.93

Document 3 → similarity = 0.12
```
The winner is:
```
Document 2
```
because its vector is closest to the question vector.

This process is called:
```
Similarity Search
```
Real example

Document contains:
```
How to reset your password
```
User searches:
```
I forgot my password
```
Notice:
```
forgot
```
and
```
reset
```
are different words.

A SQL LIKE query may miss it.

But embeddings place them close together.

So similarity search finds:

```
How to reset your password
```
even though the words don't match exactly.

That's why it's called:

```
Semantic Search
```
(search by meaning)

#### What is actually being compared?
You may wonder:
```
Are we comparing words?
```
No.

We compare vectors.

```
Question
↓
Embedding
↓
Vector Q

Document
↓
Embedding
↓
Vector D
```
Then mathematics calculates:
```
How close is Q to D?
```
The closer they are, the more relevant the document is.

#### Where Top-K comes in
Suppose the scores are:
```
Document 1 = 0.95

Document 2 = 0.91

Document 3 = 0.82

Document 4 = 0.15
```
If:
```
k = 2
```
return:
```
Document 1
Document 2
```
These are the two most similar documents.

This is called:
```
Top-K Similarity Search
```
You'll hear this term constantly in RAG.
#### The entire flow
```
User Question
       ↓
Embedding Model
       ↓
Question Vector
       ↓
Vector Database
       ↓
Similarity Search
       ↓
Top-K Documents
       ↓
LLM
       ↓
Answer
```
The most important thing to remember is:

***Similarity search does not understand English. It only compares vectors and returns the vectors that are mathematically closest to the query vector.***

###  Cosine Similarity
This is the most common similarity metric.
#### Purpose
Given two vectors:
```
A

B
```
How similar are they?

Cosine similarity gives a score.

Range
```
1.0
=
Very Similar

0
=
Unrelated

-1
=
Opposite
```
Example
```
dog

puppy

Cosine Similarity

0.98
```
Very similar.
```
dog

car

Cosine Similarity

0.15
```
What you need to remember
```
Higher score

=

More similar meaning
```
That's enough for now.

### Vector Database
Now imagine:
```
1 million documents
```
Each document has an embedding.

Example:
```
Doc1

[0.1, 0.5, ...]

Doc2

[0.7, 0.8, ...]

Doc3

[0.3, 0.2, ...]
```
Where do we store them?

A vector database.

#### What is a Vector DB?
A database optimized for:
```
Store vectors

Search vectors

Compare vectors
```
* Examples:
* Pinecone
* Weaviate
* Qdrant
* Milvus

Traditional DB:
```
SELECT *
FROM users
WHERE email='abc@gmail.com'
```
Exact match.

Vector DB:
```
Find documents
similar to this vector
```
Not exact match.

Meaning match.

### Similarity Search in Vector DB
Suppose the user asks:
```
How can I reset my password?
```
Convert query to embedding:
```
Question

↓

[0.82, 0.55, ...]
```
Vector DB compares it against every document.

Results:
```
Password Reset Guide

0.95

Authentication Guide

0.89

Docker Guide

0.12
```
Sort descending:
```
0.95
0.89
0.12
```
Most relevant first.

### Top-K Retrieval
K means:
```
Number of results
```
Example:
```
K = 3
```
Return:
```
Result 1
Result 2
Result 3
```
Pipeline:
```
User Question
↓
Embedding
↓
Vector Search
↓
Similarity Scores
↓
Top-K Results
↓
LLM
```

### Chunking
A common beginner question:
***Why not embed the whole PDF?***
Because:
```
PDF

=
100 pages
```
Embedding models have limits.

And retrieval becomes poor.

Instead:
```
PDF
↓
Chunking
↓
Small pieces
```
Example:
```
Chunk 1
Chunk 2
Chunk 3
Chunk 4
```
Each chunk gets its own embedding.

Now when user asks:
```
How to reset password?
```
The system retrieves:
```
Chunk 37
```
instead of the whole document.

This is the entire RAG foundation:
```
Documents
↓
Chunking
↓
Embeddings
↓
Vectors
↓
Vector Database
↓
Similarity Search
↓
Top-K Retrieval
↓
LLM
↓
Answer
```
