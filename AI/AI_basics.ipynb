### Artificial Intelligence basics

```
Artificial Intelligence (AI)
│
├── Machine Learning (ML)
│   │
│   └── Deep Learning (DL)
│       │
│       └── Generative AI (GenAI)
│
└── Rule-Based AI Systems
```

#### 1. Artificial Intelligence (AI)
AI is the broad field of making machines perform tasks that normally require human intelligence.

***Examples:***
* Understanding language
* Recognizing images
* Playing chess
* Making decisions
* Planning actions

**Traditional AI Example**

Suppose you build a loan approval system:
```
if salary > 50000 and credit_score > 700:
    approve_loan()
else:
    reject_loan()
```
This is AI because the machine is making a decision, but it is **not Machine Learning**.

**Key Idea**

AI = Any technique that enables a machine to mimic intelligent behavior.

#### 2. Machine Learning (ML)
Machine Learning is a subset of AI.

Instead of writing all the rules manually, we let the computer learn patterns from data.

**Traditional Programming**
```
Rules + Data
      ↓
   Output
```

***Example:***
```
if email contains "win money":
    spam = True
```
You write the rule.


**Machine Learning**
```
Data + Correct Answers
         ↓
      Model
         ↓
 Learns Rules
```
***Example:***
You give:
```
Email 1 → Spam
Email 2 → Not Spam
Email 3 → Spam
...
```
The algorithm learns the spam-detection rules itself.

***Typical ML Use Cases***
* Spam detection
* Fraud detection
* Recommendation systems
* Price prediction
* Customer churn prediction

#### 3. Deep Learning (DL)
Deep Learning is a subset of Machine Learning.

It uses artificial neural networks inspired by the human brain.

**Traditional ML**

You manually choose features.

***Example:***

Predict house price using:

```
Area
Bedrooms
Location
Age
```

Model:

```
Features → ML Algorithm → Prediction
```

### Deep Learning
The model automatically learns important features.

```
Raw Data
   ↓
Neural Network
   ↓
Prediction
```

***Example:***

For image recognition:

***Traditional ML:***
```
Image
  ↓
Extract edges
Extract colors
Extract shapes
  ↓
Model
```
***Deep Learning:***
```
Image
  ↓
Neural Network
  ↓
Dog
```
The network learns everything itself.

**Why Deep Learning Became Popular**

Because it works extremely well for:
* Images
* Speech
* Video
* Natural Language

***Examples:***
* Face recognition
* Voice assistants
* Self-driving cars
* ChatGPT-like systems

#### 4. Generative AI (GenAI)
Generative AI is a subset of Deep Learning.

Unlike traditional ML, which predicts or classifies, Generative AI creates new content.
***Traditional ML:***
Input:
```
Email
```
Output:
```
Spam / Not Spam
```
***Generative AI***

Input:
```
Write a resignation email
```
Output:
```
A completely new email generated by the model
```
The model creates content that did not previously exist.

***What GenAI Can Generate?***
* Text
* Images
* Audio
* Video
* Code

**Examples:**

* ChatGPT → text and code
* GitHub Copilot → code generation
* Midjourney → images
* Sora → videos

#### What is an LLM?

Examples:

OpenAI GPT
Anthropic Claude
Google Gemini
Meta Llama

**Understand:**

```
Input text
↓
Tokenization
↓
Prediction
↓
Output text
```


#### Step 1: Input Text
Suppose the user enters:
```
What is Django?
```
For us, this is plain text.

For the computer, it's just a sequence of characters.
```
W h a t   i s   D j a n g o ?
```
The model cannot directly work with text.

It needs numbers.

So the first step is tokenization.

#### Step 2: Tokenization
A tokenizer breaks text into smaller pieces called tokens.

A token is not necessarily:
* a word
* a character
* a sentence
It's just a chunk of text.

***Example:***
```
What is Django?
```
might become:
```
["What", " is", " D", "jango", "?"]
```
or
```
["What", " is", " Django", "?"]
```
depending on the tokenizer.
#### Why tokenize?
Because computers process numbers.

Each token gets a unique ID.

***Example:***
```
"What"   → 321
" is"    → 492
" Django"→ 8912
"?"      → 30
```
Now the text becomes:
```
[321, 492, 8912, 30]
```
This is what the model receives.

Not words.

Not sentences.

Just token IDs.

#### Step 3: Convert Tokens into Vectors
The model still can't do much with:.
```
[321, 492, 8912, 30]
```
So each token ID is converted into a vector.

***Example:***
```
321
↓
[0.12, -0.44, 0.91, ...]
```
```
492
↓
[0.55, 0.81, -0.21, ...]
```
These vectors may have thousands of dimensions.

The vector captures relationships and patterns learned during training.

#### Why vectors?
Because mathematics works on numbers.

The model performs calculations on vectors.

Not on words.

### Step 4: Prediction
This is the heart of the LLM.

The model looks at:
```
What is Django?
```
and asks:

*"What token is most likely to come next?"*

Not:

*"What is the meaning of Django?"*

Not:

*"How can I help the user?"*

Just:

*"What token should come next?"*

Suppose the model starts generating:
```
Django
```

Now the context becomes:
```
What is Django? Django
```

Then it predicts the next token.

Maybe
```
is
```
Now:
```
What is Django? Django is
```
Then:
```
a
```
Now:

```
What is Django? Django is a
```
Then:
```
high-level
```
and so on.

#### Important Insight
The model generates one token at a time.

Not the entire answer at once.

Like this:
```
Django
↓
is
↓
a
↓
high-level
↓
Python
↓
web
↓
framework
```
Thousands of times per second.

#### Why does it seem intelligent?
Because the model was trained on enormous amounts of text.

During training it learned patterns like:
```
Django is a Python web framework
```
appears frequently.

So it becomes very good at predicting what comes next.

#### Step 5: Output Tokens

After prediction, the model has produced token IDs.

Example:

```
[8912, 112, 45, 918, 22]
```

These IDs mean nothing to humans.

#### Step 6: Detokenization
The tokenizer performs the reverse operation.
```
8912 → Django
112  → is
45   → a
918  → Python
22   → framework
```
Result:
```
Django is a Python framework.
```
Now humans can read it.

### Complete Flow
```
User Types:
"What is Django?"

        ↓

Tokenizer

        ↓

[321, 492, 8912, 30]

        ↓

Embeddings (vectors)

        ↓

LLM

        ↓

Predict next token

        ↓

Predict next token

        ↓

Predict next token

        ↓

Generated token IDs

        ↓

Detokenizer

        ↓

"Django is a high-level Python web framework."
```

### Context Window
#### What is it?
Think of talking to someone who can only remember the last part of a conversation.

***Example:***
You tell someone:

*My dog's name is Bruno.*

Then you continue talking for hours about movies, food, travel, and sports.

Much later you ask:
*What's my dog's name?*

If the person still remembers "Bruno," then that information is still inside their memory window.

If they've forgotten because too much conversation happened afterward, it's outside their memory window.

**For an LLM**

The context window is:

*The maximum amount of text the model can "see" at one time.*

**That includes:**

* Your current question
* Previous messages
* Uploaded documents
* System instructions

Once the window fills up, older information may be removed.

**Tiny Example**

Context window can hold:
```
A B C D E
```
Current memory:
```
A B C D E
```
New text arrives:
```
F
```
Now:
```
B C D E F
```
"A" falls out.

The model can no longer see it.

### Inference
Inference is simply:

*The process of generating an answer.*

The process of generating an answer.

***Example***

Imagine a sentence:

*The sun rises in the _____*

You immediately think:

*east*

Your brain is performing a prediction.

An LLM does something similar.

Input:
```
The sun rises in the
```
Inference:
```
Model predicts next word
```
Output:
```
east
```
When ChatGPT answers your question, it is doing inference.

### Temperature
Temperature controls creativity versus predictability.

**Example**

Someone asks:

*Name a color*

With very low temperature:
```
Blue
```
With medium temperature:
```
Blue
Green
Red
```
some variation.

With high temperature:
```
Turquoise
Magenta
Chartreuse
Amber
```

**Rule of thumb**

Low temperature:

* Consistent
* Predictable
* Better for factual tasks

High temperature:

* Creative
* Diverse
* Better for storytelling

#### Top-K
Top-K is a filtering method.

Before choosing the next word, the model may know thousands of possible words.

Top-K says:

*Only consider the K most likely options.*

***Example***

Suppose the model predicts:

| Word      | Probability |
| --------- | ----------- |
| east      | 50%         |
| west      | 20%         |
| north     | 10%         |
| south     | 5%          |
| banana    | 0.1%        |
| spaceship | 0.01%       |

If:
```
Top-K = 3
```
Keep only:

| Word  |
| ----- |
| east  |
| west  |
| north |

Everything else is discarded.

Then the model chooses from only those three.

Intuition

Top-K answers:

*How many candidate words are allowed into the final selection pool?*

#### Top-P

Top-P is similar, but instead of a fixed number of words, it uses a probability threshold.
| Word   | Probability |
| ------ | ----------- |
| east   | 50%         |
| west   | 20%         |
| north  | 15%         |
| south  | 10%         |
| banana | 5%          |

Total:
```
100%
```
If:
```
Top-P = 0.8
```
Keep adding words from highest probability downward until reaching 80%.

Step 1:
```
east = 50%
```
Total:
```
50%
```

Step 2:

```
west = 20%
```

Total:
```
70%
```
Step 3:
```
north = 15%
```
Total:
```
85%
```
Stop.

Allowed words:
```
east
west
north
```
Discard:
```
south
banana
```
Another Example

If one word is overwhelmingly likely:

| Word  | Probability |
| ----- | ----------- |
| east  | 90%         |
| west  | 5%          |
| north | 3%          |
| south | 2%          |

Top-P = 0.8
```
east
```
is kept.

#### How they work together
```
Your Prompt
     ↓
Context Window
(what the model can see)
     ↓
Inference
(calculate possible next words)
     ↓

east 50%
west 20%
north 15%
south 10%
...

     ↓
Top-K / Top-P
(filter candidates)
     ↓
Temperature
(adjust randomness)
     ↓
Pick Next Word
     ↓
Repeat thousands of times
     ↓
Final Response
```
