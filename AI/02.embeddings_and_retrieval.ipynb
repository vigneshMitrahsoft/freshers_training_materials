### Embeddings
The Problem

Computers understand numbers.

Humans understand words.

```
Human:
"dog"

Computer:
???
```
An LLM cannot directly work with the meaning of the word "dog".

It needs a numerical representation.

What is an Embedding?

An embedding is a list of numbers that represents the meaning of a piece of text.

```
dog

↓

[0.12, -0.43, 0.91, 0.22, ...]
```
This list is called a vector.

#### Why do we need Embeddings?

Suppose we have:
```
dog
puppy
car
```
Humans know:
```
dog ≈ puppy

dog ≠ car
```
We want computers to know that too.

Embeddings make semantically similar words have similar vectors.

Example:
```
dog
↓

[0.9, 0.8, 0.7]

puppy
↓

[0.88, 0.81, 0.69]

car
↓

[0.1, 0.2, 0.9]
```
Notice:
```
dog and puppy

are close together

dog and car

are far apart
```
Except:

A hash only identifies.

An embedding captures meaning.
```
dog
↓

Embedding

↓

Meaning preserved
```

### Important: The Embedding Model is NOT the LLM

This is one of the most common beginner confusions, so let's clear it up early.

There are **two different kinds of models**, and they do different jobs.

```
Embedding Model
    input:   text
    output:  a vector  [0.12, -0.44, ...]
    job:     represent MEANING as numbers
```

```
LLM (e.g. GPT, Claude, Gemini)
    input:   text
    output:  more text
    job:     GENERATE the next tokens
```

#### Side by side

| | Embedding Model | LLM |
| --- | --- | --- |
| Gives you | A vector | Text |
| Used for | Search / similarity | Answering / writing |
| Example | `text-embedding-3-small` | `gpt-4` |

#### How to remember it

```
Embedding model  →  turns text INTO numbers (for searching)

LLM              →  turns a prompt INTO an answer (for generating)
```

In a RAG system you use **both**:

```
Embedding model  →  find the right chunks

LLM              →  write the final answer using those chunks
```

They are separate models, often from different providers, called at different steps.

### Vector
A vector is just:
```
[1, 2, 3]
```
or
```
[100]
```
That's it.

A vector is simply an ordered list of numbers.

Imagine you create your own mapping:
```
words = {
    "dog": 1,
    "cat": 2,
    "python": 3
}
```
Then:
```
"dog"
```
becomes
```
1
```
and
```
"python"
```
becomes
```
3
```
This is already text → number.

Embeddings are a smarter version

Instead of:
```
"dog" -> 1
"cat" -> 2
"car" -> 3
```
an embedding model does:
```
"dog" -> [0.8, 0.2, 0.9]
"puppy" -> [0.7, 0.3, 0.8]
"car" -> [0.1, 0.9, 0.2]
```
Notice something?
```
dog
puppy
```
have similar numbers.

Because they have similar meanings.

###  Cosine Similarity
This is the most common similarity metric.
#### Purpose
Given two vectors:
```
A

B
```
How similar are they?

Cosine similarity gives a score.

Range
```
1.0
=
Very Similar

0
=
Unrelated

-1
=
Opposite
```
Example
```
dog

puppy

Cosine Similarity

0.98
```
Very similar.
```
dog

car

Cosine Similarity

0.15
```
What you need to remember
```
Higher score

=

More similar meaning
```
That's enough for now.

### Similarity Search
Imagine you're building a Django application.

You store documents:
```
Document 1:
Django is a Python web framework.

Document 2:
FastAPI is used for building APIs.

Document 3:
Dogs are loyal pets.
```
Before storing them, you generate embeddings.
```
Document 1
↓
Embedding Model
↓
[0.81, 0.22, 0.44, ...]
```
```
Document 2
↓
Embedding Model
↓
[0.79, 0.18, 0.41, ...]
```
```
Document 3
↓
Embedding Model
↓
[0.12, 0.91, 0.73, ...]
```
User asks a question
```
How do I create REST APIs in Python?
```
We also convert this question into a vector.
```
Question
↓
Embedding Model
↓
[0.80, 0.20, 0.43, ...]
```
Now we have:
```
Question Vector
```
and
```
Document Vectors
```
#### Similarity Search
Now the system asks:

```
document vector is most similar to the question vector?
```
Something like:
```
Question Vector
      ↓

Document 1 → similarity = 0.85

Document 2 → similarity = 0.93

Document 3 → similarity = 0.12
```
The winner is:
```
Document 2
```
because its vector is closest to the question vector.

This process is called:
```
Similarity Search
```
Real example

Document contains:
```
How to reset your password
```
User searches:
```
I forgot my password
```
Notice:
```
forgot
```
and
```
reset
```
are different words.

A SQL LIKE query may miss it.

But embeddings place them close together.

So similarity search finds:

```
How to reset your password
```
even though the words don't match exactly.

That's why it's called:

```
Semantic Search
```
(search by meaning)

#### What is actually being compared?
You may wonder:
```
Are we comparing words?
```
No.

We compare vectors.

```
Question
↓
Embedding
↓
Vector Q

Document
↓
Embedding
↓
Vector D
```
Then mathematics calculates:
```
How close is Q to D?
```
The closer they are, the more relevant the document is.

#### Where Top-K comes in
Suppose the scores are:
```
Document 1 = 0.95

Document 2 = 0.91

Document 3 = 0.82

Document 4 = 0.15
```
If:
```
k = 2
```
return:
```
Document 1
Document 2
```
These are the two most similar documents.

This is called:
```
Top-K Similarity Search
```
You'll hear this term constantly in RAG.
#### The entire flow
```
User Question
       ↓
Embedding Model
       ↓
Question Vector
       ↓
Vector Database
       ↓
Similarity Search
       ↓
Top-K Documents
       ↓
LLM
       ↓
Answer
```
The most important thing to remember is:

***Similarity search does not understand English. It only compares vectors and returns the vectors that are mathematically closest to the query vector.***

Now imagine:
```
1 million documents
```
Each document has an embedding.

Example:
```
Doc1

[0.1, 0.5, ...]

Doc2

[0.7, 0.8, ...]

Doc3

[0.3, 0.2, ...]
```
Where do we store them?

A vector database.

A database optimized for:
```
Store vectors

Search vectors

Compare vectors
```
* Examples:
* Pinecone
* Weaviate
* Qdrant
* Milvus

Traditional DB:
```
SELECT *
FROM users
WHERE email='abc@gmail.com'
```
Exact match.

Vector DB:
```
Find documents
similar to this vector
```
Not exact match.

Meaning match.

### Similarity Search in Vector DB
Suppose the user asks:
```
How can I reset my password?
```
Convert query to embedding:
```
Question

↓

[0.82, 0.55, ...]
```
Vector DB compares it against every document.

Results:
```
Password Reset Guide

0.95

Authentication Guide

0.89

Docker Guide

0.12
```
Sort descending:
```
0.95
0.89
0.12
```
Most relevant first.

### Top-K Retrieval
K means:
```
Number of results
```
Example:
```
K = 3
```
Return:
```
Result 1
Result 2
Result 3
```
Pipeline:
```
User Question
↓
Embedding
↓
Vector Search
↓
Similarity Scores
↓
Top-K Results
↓
LLM
```

### Semantic Search
Semantic search is a type of search that finds results based on meaning rather than exact keywords.

Example:

User searches:
```
How can I create REST APIs?
```
Document contains:
```
Building web services using FastAPI
```
No exact words match.

But the meaning is similar.

Semantic search still finds it.

#### How does Semantic Search work?
Usually through embeddings + similarity search.
```
Text
↓
Embeddings
↓
Vectors
↓
Similarity Search
↓
Relevant Results
```

#### The Relationship
```
Semantic Search
        uses
          ↓
Similarity Search
```

#### Analogy
Suppose you're building a Django app.

Goal
```
User Login
```
Technique
```
Password Hash Comparison
```
Similarly:

Goal
```
Semantic Search
```
Technique
```
Similarity Search
```

#### Example
Suppose we have documents:
```
Doc1:
How to build REST APIs with FastAPI

Doc2:
How to train a neural network

Doc3:
Python database optimization
```
User asks:
```
How do I create backend APIs?
```
Keyword Search

Looks for:
```
backend
APIs
create
```
Maybe finds nothing.
#### Semantic Search
Convert everything to embeddings.
```
Query
↓
Embedding

Doc1
↓
Embedding

Doc2
↓
Embedding

Doc3
↓
Embedding
```
Run similarity search.

Results:
```
Doc1 = 0.93
Doc3 = 0.51
Doc2 = 0.12
```
Return:
```
Doc1
```

### Hybrid Search

So far we have seen two ways to search:

```
Keyword Search   →  matches exact words
Semantic Search  →  matches meaning (embeddings)
```

Each one is good at something, and bad at something else.

Hybrid search simply means:

```
Use BOTH at the same time
and combine their scores
```

#### Why do we need it?

Semantic search is powerful, but it is not perfect.

Consider this query:
```
What is error code E-4042?
```

Semantic search may struggle because:
```
"E-4042" has no real "meaning"
It is just an exact token
```

A keyword search is perfect for this, because it matches the exact string `E-4042`.

Now consider this query:
```
How do I stop my account from being charged?
```

A keyword search may fail, because the document says:
```
Cancel your subscription
```
No words match. But semantic search understands the meaning.

So:

| Query type | Keyword Search | Semantic Search |
| --- | --- | --- |
| Exact IDs, codes, names (`E-4042`, `SKU-19`) | Excellent | Weak |
| Meaning / paraphrase ("forgot password") | Weak | Excellent |
| Rare / typo'd words | Good | Sometimes weak |

Hybrid search gives you the **best of both**.

#### Keyword Search (a quick refresher)

Keyword search is usually done with an algorithm called **BM25** (or older TF-IDF).

```
BM25 looks at:
  - Does the word appear in the document?
  - How often?
  - How rare is that word overall?
```

It does NOT understand meaning. It only counts words.

```
"reset password"  →  finds docs containing "reset" and "password"
```

#### How Hybrid Search works

You run both searches, then merge the results.

```
              User Query
                  │
        ┌─────────┴─────────┐
        ▼                   ▼
  Keyword Search       Semantic Search
   (BM25)               (Embeddings)
        │                   │
   keyword scores      vector scores
        └─────────┬─────────┘
                  ▼
          Combine the scores
                  ▼
            Final Ranking
                  ▼
            Top-K Results
```

#### Combining the scores

The two searches produce scores on **different scales**:

```
BM25 score      →  e.g. 0  to  18.4   (unbounded)
Cosine score    →  e.g. 0  to  1.0    (bounded)
```

You cannot just add them directly. There are two common ways to combine them.

**1. Weighted sum (after normalizing)**

Bring both scores to the same range (0–1), then blend with a weight `alpha`:

```
final = alpha * semantic_score  +  (1 - alpha) * keyword_score
```

```
alpha = 1.0  →  pure semantic search
alpha = 0.0  →  pure keyword search
alpha = 0.5  →  equal mix
```

**2. Reciprocal Rank Fusion (RRF)**

Instead of using the raw scores, use the **rank** (position) of each document in each list. This avoids the "different scales" problem entirely.

```
RRF_score(doc) = Σ   1 / (k + rank_in_list)
                 lists
```

```
k is a constant (commonly 60)
rank starts at 1 for the top result
```

A document that ranks high in BOTH lists gets the best combined score.

#### Small RRF example

Two result lists for the same query (using `k = 60`):

```
Keyword results:        Semantic results:
1. DocA                 1. DocB
2. DocB                 2. DocC
3. DocC                 3. DocA
```

Score each document by summing `1 / (k + rank)`:

```
DocA = 1/(60+1) + 1/(60+3) = 0.0164 + 0.0159 = 0.0323
DocB = 1/(60+2) + 1/(60+1) = 0.0161 + 0.0164 = 0.0325
DocC = 1/(60+3) + 1/(60+2) = 0.0159 + 0.0161 = 0.0320
```

Final ranking:
```
1. DocB   (strong in both)
2. DocA
3. DocC
```

`DocB` wins because it scored well in **both** searches, even though it was not #1 in the keyword list.

#### When should you use Hybrid Search?

```
Use Hybrid Search when your data has BOTH:
  - natural language  (descriptions, questions, docs)
  - exact tokens      (codes, IDs, names, acronyms)
```

This is very common in real products:
```
E-commerce   →  "comfy running shoes" + "SKU-2291"
Support docs →  "can't log in" + "error 0x80070005"
Code search  →  "function that sorts a list" + "def quicksort"
```

#### Key things to remember

```
Hybrid Search = Keyword Search + Semantic Search
```

- Keyword search is great for **exact** matches (IDs, codes, rare words).
- Semantic search is great for **meaning** (paraphrases, synonyms).
- Their scores are on different scales, so you must **normalize** or use **RRF** before combining.
- Most modern vector databases (Pinecone, Weaviate, Qdrant, Milvus) support hybrid search out of the box.
