## What is a Website?


A website is mainly used to display information to users.

Examples:
- Company websites
- Blogs
- News websites
- Portfolio websites

Characteristics:
- Mostly static content
- Less user interaction
- Usually no login system
- Mainly informational

Technologies:
- HTML
- CSS
- JavaScript


## What is a Web Application?


A web application allows users to perform actions and interact with the system.

Examples:
- Gmail
- Instagram
- Google Docs
- Online Banking
- E-commerce websites

Characteristics:
- Dynamic content
- User authentication
- Database integration
- High user interaction
- Business logic processing


## Website vs Web Application


| Feature | Website | Web Application |
|---|---|---|
| Purpose | Show information | Perform tasks |
| Interaction | Low | High |
| Login | Usually No | Usually Yes |
| Database | Optional | Common |
| Dynamic Content | Less | More |


## How Internet Works


Flow:

Browser → Internet → Server → Response → Browser

Example:
1. User enters a URL
2. Browser sends request
3. Server processes request
4. Server sends response
5. Browser displays content

# Website Request-Response Workflow

#### 1. Browser
The user types a URL (like `google.com`) into the browser.

The browser sends a request to find that website.

#### 2. DNS (Domain Name System)
DNS works like a phonebook.

It converts the domain name (`google.com`) into an IP address (like `142.250.190.78`) that computers use to communicate.

#### 3. Web Server
The request goes to the server hosting the website.

The server processes the request:
- opening webpages
- fetching files
- running backend logic

#### 4. Database Server
If the webpage needs stored information:
- user profile
- search results
- posts
- login details

the web server asks the database server for that data.

The database sends the required data back to the web server.

#### 5. Response back to Browser
The server sends:
- HTML
- CSS
- JavaScript
- images
- data

back to the browser.

The browser renders everything, and you see the webpage.

## Frontend vs Backend


Frontend:
- What users see
- UI/UX
- Built using HTML, CSS, JavaScript

Backend:
- Runs on server
- Handles business logic
- Connects database
- Authentication
- APIs

Popular backend languages:
- Python
- Node.js
- Java
- PHP

Popular Python Frameworks:

## Important Terms


1. Server
    - A computer that runs backend code.

2. Client
    - User's browser or device.

3. Database
    - Stores application data.

4. API
    - Communication bridge between frontend and backend.

5. HTTP Methods
- GET → Fetch data
- POST → Send data
- PUT → Update data
- DELETE → Remove data

## Static vs Dynamic Websites


Static Website:
- Fixed content
- Same for every user

Dynamic Website:
- Content changes based on user/data
- Uses backend and databases
