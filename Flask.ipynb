**Flask**


What is Flask?
  * It was developed by **Armin Ronacher** in 2010
  * Flask is a lightweight and flexible web framework for Python
  * It is also  easier to learn because it has less base code to implement a simple web application.

**Where and why we use flask?**
  * If you are a beginner in framework and want to build a small application,You can go to flask because it is light weight, simple so you can easily develop an application
  * You want to make application for small organization for limited amount to operation , so you can go with flask , it is bit faster then django form smalll amout of task and less amount of code

**Advantages of Flask:**

  * **Lightweight**: Flask is a lightweight framework with a simple and intuitive design, making it easy to learn and use.

  * **Flexibility**: Flask offers a high level of flexibility, allowing developers to choose libraries and tools that suit their needs.

  * **Rapid Development**: Flask provides the ability for rapid development and is suitable for building small projects

**Disadvantage of Flask:**
  * **Lack of built-in features:** Compared to other frameworks, Flask has fewer built-in features and may require additional extension libraries.

  * **Lack of standardization:** Flask project structures can vary among developers, lacking a unified standardization.

**What is Flask-sqlalchemy:**
  * Flask-SQLAlchemy is a helper library (also known as a Flask "extension") that makes it much easier to use the powerful SQLAlchemy library within your Flask web applications. Think of it as a bridge that simplifies communication between your Flask application and your database.

**Why we use flask-sqlalchemy instead of sqlalchemy:**
  * **Simple database setup**: It provides a straightforward way to configure your database connection within Flask. Instead of manually creating engine objects and setting up sessions with SQLAlchemy, Flask-SQLAlchemy takes care of it for you.
  * It simplifies specifying the database URL (e.g., the address and credentials for connecting to your database).
  * **Model definition:** It provides a straightforward way to configure your database connection within Flask. Instead of manually creating engine objects and setting up sessions with SQLAlchemy, Flask-SQLAlchemy takes care of it for you.
  * It simplifies specifying the database URL (e.g., the address and credentials for connecting to your database).
  * **Simplified Querying:** Flask-SQLAlchemy adds a query attribute to your models that makes it easier to perform database queries. You don't have to write raw SQL statements.
  * You can use methods like filter_by(), first(), all(), and get() to retrieve data from your database.
